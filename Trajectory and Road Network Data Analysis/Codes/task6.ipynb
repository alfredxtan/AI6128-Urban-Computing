```python
import os
# import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import math


def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0]) ** 2 + (p1[1] - p2[1]) ** 2) * 100

def list2string(L):
    string = str(L)
    return string.replace(' ', '')


def make_folder(folder):
    if not os.path.exists(folder):
        os.makedirs(folder)


folder = './task6_results'
make_folder(folder)

train1500 = pd.read_csv('./data/train-1500.csv')

consecutive_dis = []
```

```python
# counting...
for t_number in range(1500):
    gps_points = eval(train1500['POLYLINE'][t_number])
    for i in range(len(gps_points)-1):
        pre = gps_points[i]
        post = gps_points[i+1]
        cur_dis = distance(pre, post)
        consecutive_dis.append(cur_dis)
#%%



plt.hist(x = consecutive_dis, 
    bins = 20,
    color = 'lightgreen',
    edgecolor = 'black',
    range=(0,2)
    )

plt.savefig(folder+'/distribution.png', dpi=240)

```

```python

#outlier modifying
threshold_min = 0.5
threshold_max = 2
threshold_step = 0.2

# threshold_min = 0.01
# threshold_max = 0.02
# threshold_step = 0.002

for t in range(1500):
    gps_points = eval(train1500['POLYLINE'][t])
    assert isinstance(train1500['POLYLINE'][t], str)

    if len(gps_points) < 2: # if there are less than GPS points in this trajectory
        modified_points = gps_points
    else:
        modified_points = gps_points[:1]
        last_point = gps_points[0]
        threshold = threshold_min
        for i in range(1, len(gps_points)):
            next_point = gps_points[i]
            if distance(last_point, next_point) < threshold: # a valid point
                modified_points.append(next_point)
                last_point = next_point
                threshold = threshold_min
            else:
                threshold += threshold_step
                threshold = min(threshold, threshold_max)
        train1500['POLYLINE'][t] = list2string(modified_points)

train1500.to_csv(folder+'/modified_train1500.csv',index=0)
    
```
